```python
#CHATGPT4 - LYRICS (CLOSED)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

df = pd.read_csv('similarity scores_GPT4 - similarity scores_GPT4.tsv', delimiter='\t')


unique_labels = sorted(df['Gold'].unique())

# Calculate Precision, Recall, and F1 scores and print a classification report
report = classification_report(df['Gold'], df['L1'], target_names=unique_labels)

print("Classification Report:")
print(report)

confusion_matrix = confusion_matrix(df['Gold'], df['L1'], labels=unique_labels)

# Create a heatmap for the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=unique_labels, yticklabels=unique_labels)
plt.title("Confusion Matrix")
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

df = pd.read_csv('similarity scores_GPT3.5 - similarity scores_GPT3.5.tsv', delimiter='\t')

# Sort class labels
unique_labels = sorted(df['Gold'].unique())

# Calculate Precision, Recall, and F1 scores and store them in a DataFrame
report = classification_report(df['Gold'], df['L1'], target_names=unique_labels, output_dict=True)
report_df = pd.DataFrame(report).T  # Transpose the report to have metrics as columns

# Round Precision, Recall, and F1 scores to 2 decimal places
report_df[['precision', 'recall', 'f1-score']] = report_df[['precision', 'recall', 'f1-score']].round(2)

# Save the Precision, Recall, and F1 scores as a TSV file
report_df.to_csv('3.5L1_precision_recall_f1_scores.tsv', sep='\t', float_format='%.2f')

# Generate a confusion matrix with sorted labels
confusion_matrix = confusion_matrix(df['Gold'], df['L1'], labels=unique_labels)

# Create a heatmap for the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=unique_labels, yticklabels=unique_labels)
plt.title("Confusion Matrix")
plt.show()

```
